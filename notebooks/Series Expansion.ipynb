```python
from sympy import *
x, y, z = symbols('x y z')
init_printing(use_unicode=True)
import numpy as np
import matplotlib.pyplot as plt
```

```python
expr = Integral(log(x)**2, x)
expr
expr.doit()
```

```python
expr = exp(sin(x))
expr1 = expr.series(x, 2, 4)
expr1
```

The $O(x^4)$ term at the end represents the Landau order term at $x=0$
 (not to be confused with big O notation used in computer science, which generally represents the Landau order term at $x$
 where $x\rightarrow\infty$
). It means that all x terms with power greater than or equal to $x^4$
 are omitted. Order terms can be created and manipulated outside of series. They automatically absorb higher order terms.



```python
t = np.arange(0.0, 10.0, 0.05)
s1 = np.exp(np.sin(t))

expr1 =expr1.removeO()
s2 = lambdify(x, expr1)

fig, ax = plt.subplots()
# plt.subplot(1, 2, 1)
# plt.plot(t, s1)

# plt.subplot(1, 2, 2)
# plt.plot(t,s2(t))


ax.plot(t, s1)
ax.plot(t,s2(t))
ax.set(xlim=(0, 2), ylim=(0, 5))
```
